```python
dataset = pd.read_csv('netflix_data.csv', header = None , encoding="latin1")
transactions = []

for i in range(0, 7466):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)]) 
```

```python
print(dataset)
```

```python
from apyori import apriori
anmol_intelligence = apriori (transactions = transactions, min_support = 0.002 , min_confidence = 0.2 , min_lift = 1 , min_length = 2)
```

```python
results = list(anmol_intelligence)
results
```

```python
def inspect(results):
    rows = []
    
    for result in results:
        for stat in result.ordered_statistics:
            
            if len(stat.items_base) > 0 and len(stat.items_add) > 0:
                
                movie1 = list(stat.items_base)[0]
                movie2 = list(stat.items_add)[0]
                support = result.support
                confidence = stat.confidence
                lift = stat.lift
                
                rows.append((movie1, movie2, support, confidence, lift))
    
    return rows
```

```python
DataFrame_intelligence = pd.DataFrame(
    inspect(results),
    columns=['Movie1','Movie2','Support','Confidence','Lift']
)

DataFrame_intelligence
```

```python
import pandas as pd
import pickle

# load rule table
rules = pd.read_csv("movie_rules.csv")


class MovieRecommender:

    def __init__(self, rules):
        self.rules = rules

    def recommend(self, movie):

        result = self.rules[
            self.rules["Movie1"].str.lower() == movie.lower()
        ][["Movie2","Confidence","Lift"]]

        result = result.drop_duplicates(subset=["Movie2"])
        result = result.sort_values(by="Lift", ascending=False)

        return result.head(10)

    def relation(self, movie1, movie2):

        result = self.rules[
            (self.rules["Movie1"].str.lower() == movie1.lower()) &
            (self.rules["Movie2"].str.lower() == movie2.lower())
        ][["Support","Confidence","Lift"]]

        if result.empty:
            return "No strong relation found"

        result = result.sort_values(by="Lift", ascending=False)

        return result.head(1)


# create object
model = MovieRecommender(rules)

# save model
pickle.dump(model, open("movie_model.pkl","wb"))
```
